```python
import pandas as pd;
import seaborn as sns;
import numpy as np;
import matplotlib.pyplot as plt;
%matplotlib inline
```

### Remember: this is for unsupervised learning!!!

```python
from sklearn.datasets import make_blobs;
```

```python
df = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

```python
# for K means clustering:

from sklearn.cluster import KMeans;
```

```python
# NOTE: we do need to know the no. of clusters before-hand for this!
kmeans = KMeans(n_clusters = 4)
```

```python
# fit to features
kmeans.fit(df[0]);
```

```python
# To get centers of cluster!
kmeans.cluster_centers_
```

```python
# for labels!
kmeans.labels_
```

<strong>NOTE:</strong> We would be done by here if this were real data and we didn't have the labels. We wouldn't be able to compare anything to the real values.

```python
# but since this is an artificial dataset!
# Correct labels:
df[1]
```

```python
# Predicted labels:
kmeans.labels_
```

Since K means clustering is an unsupervised learning technique, we would not use these to perform predictions usually

```python
# but since we are working on an artificial dataset:

fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10, 6))
# algo data colored by predictions
ax1.set_title('K Means');
ax1.scatter (df[0][:, 0], df[0][:, 1], c = kmeans.labels_, cmap = 'rainbow');
# algo data colored by actually assigned labels/blobs
ax2.set_title('Original');
ax2.scatter (df[0][:, 0], df[0][:, 1], c = df[1], cmap = 'rainbow');
```
