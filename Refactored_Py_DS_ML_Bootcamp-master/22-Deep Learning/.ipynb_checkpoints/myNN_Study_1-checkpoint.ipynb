```python
import pandas as pd;
import numpy as np;
import seaborn as sns;
import matplotlib.pyplot as plt;
%matplotlib inline
```

```python
df = pd.read_csv('./TensorFlow_FILES/DATA/fake_reg.csv');
df.head()
```

target variable: price <br>
variables to be used for training: feature 1 and feature 2

## Exploratory Data Analysis (EDA)

```python
sns.set_style('darkgrid')
sns.pairplot(df)
```

## Train

```python
from sklearn.model_selection import train_test_split;
```

```python
## NOTE: using the .values to convert values into an array for the NN
X = df[['feature1', 'feature2']].values;
X
```

```python
## NOTE: using the .values to convert values into an array for the NN
y = df['price'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler;
```

```python
help(MinMaxScaler)
```

```python
help(StandardScaler)
```

```python
scaler = MinMaxScaler()
```

<strong>NOTE:</strong> not gonna fit X_test at all to prevent data leaking!

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train);
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test);
X_test_scaled
```

```python
print ('For training set:\nmin: ', X_train_scaled.min(), ' max: ', X_train_scaled.max(), '\n')
```

```python
print ('For test set:\nmin: ', X_test_scaled.min(), ' max: ', X_test_scaled.max(), '\n')
```

## Construct Neural Network

```python
#!conda list
```

```python
import tensorflow as tf;
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
```

```python
#help(Sequential)
INPUT_VALS = 4
n_epochs = 300
```

```python
model = Sequential();
```

```python
model.add(Dense(INPUT_VALS, activation = 'relu'));
model.add(Dense(4, activation = 'relu'));
model.add(Dense(4, activation = 'relu'));

# depending on what we are trying to predict here; 
# therefore we need a single neuron here 
model.add(Dense(1));
```

```python
model.compile(optimizer = "rmsprop", loss = 'rmse', metrics = ['accuracy'])
```

```python
model.fit(X_train_scaled, y = y_train, epochs = n_epochs, verbose = 1)
```
