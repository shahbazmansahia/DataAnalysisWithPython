```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiplots on the same canvas:
# format: plt.subplot(<numRows>, <numCols>, <plot values>)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot (y, x, 'b')
```

```python
# via an OOP method (preferred/better):
# basic idea: create figure objects --> call methods off/on figure
# NOTE: Figure is basically an empty canvas object
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title ('Set Title')
```

```python
fig = plt.figure()
# Format: <figure>.add_axes([<left>, <bottom>, <width>, <height>])
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Translation of code below: 
# <figure>.add_axes([<20% of left>, <50% of bottom>, <40% of canvas size and width>, <30% of canvas size in height>])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.15, 0.4, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot (x, y)
axes1.set_title('LARGER PLOT')

axes2.plot (y, x)
axes2.set_title('SMALLER PLOT')
```

### Figure Size, Aspect Ratio and DPI

```python
fig  = plt.figure (figsize=(3, 2));
ax = fig.add_axes([0, 0, 1, 1]);
ax.plot(x, y);
```

```python
# to make it wider; specifically 8x2 inches:
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
# Plot/graph Legends
fig = plt.figure()
ax = fig.add_axes ([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label ='X Cubed')

ax.legend()
```

```python
# Plot/graph Legends
fig = plt.figure()
ax = fig.add_axes ([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label ='X Cubed')
# we can also change the location of the legend
ax.legend(loc = 0) # 0 == "best"
```

```python
# Plot/graph Legends
fig = plt.figure()
ax = fig.add_axes ([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label ='X Cubed')
# we can also change the location of the legend
ax.legend(loc = 10) # 0 == "center"
```

```python
# Plot/graph Legends
fig = plt.figure()
ax = fig.add_axes ([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label ='X Cubed')
# we can also change the location of the legend
ax.legend(loc = (0.1, 0.1)) # passed location as legend 10% from left and bottom!
```

```python
# x and y limits
# line style

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot (x, y, lw=3, color='purple', ls='--')

# to only show plot between 0 and 1 on x axis
ax.set_xlim([0, 1])
```

```python
# x and y limits
# line style

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot (x, y, lw=3, color='purple', ls='--')

# to only show plot between 0 and 2 on y axis
ax.set_ylim([0, 2])
```
