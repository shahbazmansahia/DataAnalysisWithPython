## EDA

## Support Vector Machine Model

```python
from sklearn.model_selection import train_test_split;
```

```python
X = df_feat;
y = cancer['target'];
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101);
X_train.reset_index(drop = True);
X_test.reset_index(drop = True);
#y_train.reset_index(drop = True);
#y_test.reset_index(drop = True);
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test);
```

```python
from sklearn.metrics import classification_report, confusion_matrix;
```

```python
print (confusion_matrix(y_test, y_test_pred), "\n");
print (classification_report(y_test, y_test_pred));
```

```python
# grid search for better performance
from sklearn.model_selection import GridSearchCV;
```

```python
# dictionary for grid; keys are the parameters that go into the model.
# for SVC, the params are C, cache_size, class_weight, coef0, decision_function_shape, degree, gamma, kernel, etc.
# for SVM models: rbf == 'radial basis funtion'!!! Usually the best one to use...
# gamma for SVM models: a small gamma means a gaussian for large variance; i.e. a large gamma == (high-bias & low variance) and vice versa
param_grid = {'C':[0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]};
```

```python
# estimator here is the SVC(); the 'support vector classifier'
# verbose basically gives us a text output of our desctiption/result. the larger the no., the more detailed the output
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid_pred = grid.predict(X_test);
```

```python
print (confusion_matrix(grid_pred, y_test), "\n");
print (classification_report(grid_pred, y_test));
```

```python
# grid search is actually very nice and one of the only options for getting good C and gamma vals for an estimator!!
```
